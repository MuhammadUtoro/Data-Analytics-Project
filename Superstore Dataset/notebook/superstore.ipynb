### End to end data analysis project using dataset from kaggle. In python, I'll do data cleaning and EDA

```python
# Importing Package
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import chardet
```

```python
# We check the encoding first
with open('../data/Superstore.csv', 'rb') as f:
    result = chardet.detect(f.read())

result
```

```python
# Load the dataset
df_raw = pd.read_csv('../data/Superstore.csv', encoding='latin1')
df_raw.head(10)
```

```python
# Observing data
df_raw.info()
# We can see that we don't have any missing values.
```

```python
df_raw['Row ID']
```

```python
# Since the row ID just contan the row number, then we don't need id
df = df_raw.copy()
df = df.drop('Row ID', axis=1)
df.info()
```

```python
# Now we change the type of Order Date and Shipe Date to Datetime
for i in ['Order Date', 'Ship Date']:
    df[i] = pd.to_datetime(df[i])
df.info()
```

```python
# Now we extract the year and month from Order Date and Ship Date then insert a new column for trend analysis later on
df['Order Year'] = df['Order Date'].dt.year 
df['Ship Year'] = df['Ship Date'].dt.year
df['Order Month'] = df['Order Date'].dt.month_name()
df['Ship Month'] = df['Ship Date'].dt.month_name()
df.info()
```

```python
# Now we make all the columns name lower case and change whitespace with _
df.columns = df.columns.str.lower()
df.columns =  [col.replace(" ", "_") for col in df.columns]
```

```python
# Now we import the data into SQL and do data analysis
# Install sqlalchemy and pycopg1-binary and connect
from sqlalchemy import create_engine
```

```python
username =  "postgres"
password = "admin"
host = "localhost"
port = "5432"
db = "customer_behavior"

# Start the Postgresql
engine = create_engine(
    f"postgresql+psycopg2://{username}:{password}@{host}:{port}/{db}"
)

table_name = "customer"
df.to_sql(table_name, engine, if_exists="replace", index=False)

print(f"Data loaded to table: '{table_name}' in database; '{db}'")
```
